```python
df = pd.DataFrame(data)


df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Year'] = df['Date'].dt.year
df['DayOfYear'] = df['Date'].dt.dayofyear
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.weekday
df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)


df = pd.get_dummies(df, columns=['District'], drop_first=False)

# print(df.head(5))


# Features and target
features = ['Year', 'Month', 'Month_sin', 'Month_cos', 'DayOfYear', 'Day', 
            'Weekday'] + [col for col in df.columns if 'District_' in col]
X = df[features]
y = df['Avg_smlvl_at15cm']


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

import lightgbm as lgb
from sklearn.metrics import mean_squared_error, r2_score

model = lgb.LGBMRegressor(objective='regression', num_leaves=31, learning_rate=0.05, n_estimators=100)
model.fit(X_train, y_train)

# using
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
# print(f'MSE: {mse}, R: {r2}')


# print(f"\n\n\n\n\n\n\n\nTRAINING SCORE OF MODEL IS {model.score(X_train,y_train)*100}")
# print(f"TRAINING SCORE OF MODEL IS {model.score(X_test,y_test)*100}")


from sklearn.model_selection import GridSearchCV , RandomizedSearchCV
'''
model100 = lgb.LGBMRegressor()
num_leaves: int = 31, max_depth: int = -1, learning_rate: float = 0.1, n_estimators: int = 100
'''

estimator_grid = {"num_leaves" : [31,40,50,60,70,100], "n_estimators" : [100,200,300,400] , "learning_rate" : [0.05,0.1,0.2,0.3]}

'''
grid_sr = GridSearchCV(model , param_grid = estimator_grid , cv = 5 , verbose = 1 , n_jobs = -1)
grid_sr.fit(X_train,y_train) 


print(grid_sr.best_params_) #''learning_rate': 0.2, 'n_estimators': 400, 'num_leaves': 100
print(grid_sr.best_score_) #9542504100543147
'''

'''
random_sr = RandomizedSearchCV(estimator=model,
    param_distributions=estimator_grid,
    n_iter=10,           # Number of parameter settings sampled
    cv=5,                # 5-fold cross-validation
    verbose=1,
    n_jobs=-1,
    scoring='neg_mean_squared_error'  # or another suitable metric
)
'''
model_new = lgb.LGBMRegressor(objective='regression', 
                              learning_rate=0.2, n_estimators=400, num_leaves=100)
model_new.fit(X_train, y_train)

print(f"\n\n\n\n\n\n\n\nTRAINING SCORE OF NEW MODEL IS {model_new.score(X_train,y_train)*100}")
print(f"TESTING SCORE OF NEW MODEL IS {model_new.score(X_test,y_test)*100}")

'''
random_sr.fit(X_train, y_train)
print(random_sr.best_params_)
print(random_sr.best_score_)
'''
```
